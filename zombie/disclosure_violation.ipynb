## 1. 시가총액 요건 (가장 중요, 실시간 모니터링 필요)

**어디서:** HTS/MTS (키움, 미래에셋, 신한 등) 또는 네이버 증권, 한국거래소(KRX)

**무엇을:**

- 코스닥 기준 시총 200억 원 미만(2026.7~), 300억 원 미만(2027.1~) 기업 필터링
- 코스피 기준 300억 원 미만(2026.7~), 500억 원 미만(2027.1~)
- KRX 정보데이터시스템(data.krx.co.kr) → 종목검색 → 시가총액 정렬

**핵심 포인트:** 30거래일 연속 기준치 미달이 트리거이므로, 현재 시총이 기준치의 1.5~2배 이내인 기업은 잠재 위험군.

### 잠재 위험군

#### 코스닥 기준 (2026년 7월 시총 200억 원 적용)

**"잠재 위험군" = 현재 시총 200억~400억 원대 기업**

예를 들어 현재 시총이 **280억 원**인 코스닥 기업이 있다면:

- 200억 원까지 불과 **28% 하락**만으로 기준치 미달
- 코스닥 일간 변동성(±3~5%)을 감안하면 한 달 안에 충분히 도달 가능한 수준
- 즉 "지금은 괜찮아 보이지만 시장 악화 시 빠르게 관리종목 지정 가능"

반면 시총 **800억 원** 기업은 기준치(200억)의 4배 수준이므로 절반 토막이 나야 위험 — 단기 우려는 낮음.

#### 코스피 기준 (2026년 7월 시총 300억 원 적용)

**"잠재 위험군" = 현재 시총 300억~600억 원대 기업**

코스피 소형주 중 시총 **450억 원** 기업이라면 300억까지 **33% 하락**이면 기준치 이탈.

#### 왜 "1.5~2배"를 기준으로 잡는가?

|현재 시총|기준치|하락 여유|위험도|
|---|---|---|---|
|250억 원|200억 원|–20%|🔴 고위험|
|350억 원|200억 원|–43%|🟡 중위험|
|500억 원|200억 원|–60%|🟢 상대적 안전|

코스닥 소형주는 악재 하나에 20~30% 급락이 드물지 않기 때문에, 기준치의 **2배 이내(400억 미만)**는 현실적 위험군으로 보는 겁니다.

#### 실제로 찾으려면

KRX 정보데이터시스템(data.krx.co.kr)에서 코스닥 전체 종목을 시가총액 오름차순으로 정렬하면, 현재 **200억~400억 원 구간**에 있는 종목들이 그대로 리스트업됩니다. 2026년 3월 현재 기준으로 이 구간에 수십 개 종목이 해당될 가능성이 높습니다.

원하시면 이 기준으로 스크리닝 체크리스트를 엑셀로 만들어 드릴 수 있습니다.

```python
from pykrx import stock
import pandas as pd

ticker = "123456"  # 종목코드
df = stock.get_market_cap_by_date("20260101", "20260306", ticker)

threshold = 20_000_000_000  # 200억
df['미달'] = df['시가총액'] < threshold

# 연속 미달일 수 계산
df['연속일수'] = df['미달'] * (df.groupby(
    (df['미달'] != df['미달'].shift()).cumsum()).cumcount() + 1)

print(df[['시가총액', '미달', '연속일수']].tail(40))
```
