```python
import json
import os
import dart_fss as dart

api_key = '888e4aca3aba61a62811ba87e0fc054e2d9f6ea0'
dart.set_api_key(api_key=api_key)

# 저장된 JSON 파일에서 회사 리스트 로드
def load_corp_list_from_json(json_file='/Users/daegyunggang/Documents/workspace/indgram/notebook/corp_list.json'):
    """
    JSON 파일에서 회사 리스트를 로드하여 CorpList 객체로 반환
    """
    if not os.path.exists(json_file):
        print(f"파일이 존재하지 않습니다: {json_file}")
        print("새로 데이터를 가져옵니다...")
        return dart.get_corp_list()
    
    try:
        with open(json_file, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        # CorpList 객체 생성
        corp_list = dart.corp.CorpList()
        
        # 각 회사 정보를 Corp 객체로 변환하여 추가
        for item in data:
            corp = dart.corp.Corp(
                corp_code=item['corp_code'],
                corp_name=item['corp_name'],
                stock_code=item.get('stock_code'),  # None일 수 있음
                modify_date=item['modify_date']
            )
            corp_list._corps.append(corp)
        
        print(f"JSON 파일에서 {len(corp_list._corps):,}개 회사 로드 완료")
        return corp_list
        
    except Exception as e:
        print(f"JSON 파일 로드 실패: {e}")
        print("새로 데이터를 가져옵니다...")
        return dart.get_corp_list()

# 회사 리스트 조회 (저장된 파일 활용)
corp_list = load_corp_list_from_json()
```
