# 회사 코드 검색기
`corp_list.csv`를 로드하여 회사명으로 회사 코드를 검색합니다.

```python
import pandas as pd
import os

# 파일 경로 설정
csv_path = 'fsdata/corp_list.csv'

# CSV 파일 로드 (corp_code가 문자열로 유지되도록 dtype 설정)
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path, dtype={'corp_code': str, 'stock_code': str})
    print(f"Successfully loaded {len(df)} companies.")
else:
    print(f"Error: File not found at {csv_path}")
```

```python
def get_corp_code(corp_name):
    """
    회사명을 입력받아 해당 회사의 정보를 반환합니다.
    부분 일치(contains) 검색을 수행합니다.
    """
    if 'df' not in globals():
        return "Dataframe not loaded."
    
    result = df[df['corp_name'].str.contains(corp_name, na=False)]
    
    if result.empty:
        return f"'{corp_name}'에 대한 검색 결과가 없습니다."
    
    return result

# 검색 실행 예시
target_company = "삼성전자"  # 검색하고 싶은 회사명 입력
get_corp_code(target_company)
```
